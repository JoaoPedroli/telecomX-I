#ðŸ“Œ ExtracÃ£o

```python
import pandas as pd
import requests
import json

response = requests.get('https://raw.githubusercontent.com/alura-cursos/challenge2-data-science/main/TelecomX_Data.json')
data = json.loads(response.text)

df = pd.json_normalize(data, sep='_')

df.head()
```

```python
df['Churn'] = df['Churn'].apply(lambda x: 1 if x == 'Yes' else 0)
df['customer_Partner'] = df['customer_Partner'].apply(lambda x: 1 if x == 'Yes' else 0)
df['customer_Dependents'] = df['customer_Dependents'].apply(lambda x: 1 if x == 'Yes' else 0)
df['phone_PhoneService'] = df['phone_PhoneService'].apply(lambda x: 1 if x == 'Yes' else 0)
df['phone_MultipleLines'] = df['phone_MultipleLines'].apply(lambda x: 1 if x == 'Yes' else 0)
df['internet_DeviceProtection'] = df['internet_DeviceProtection'].apply(lambda x: 1 if x == 'Yes' else 0)
df['internet_TechSupport'] = df['internet_TechSupport'].apply(lambda x: 1 if x == 'Yes' else 0)
df['internet_StreamingTV'] = df['internet_StreamingTV'].apply(lambda x: 1 if x == 'Yes' else 0)
df['internet_StreamingMovies'] = df['internet_StreamingMovies'].apply(lambda x: 1 if x == 'Yes' else 0)
df['account_PaperlessBilling'] = df['account_PaperlessBilling'].apply(lambda x: 1 if x == 'Yes' else 0)
df.head()
```

#ðŸ”§ TransformaÃ§Ã£o

```python
df.info()
```

```python
df['account_Charges_Total'] = pd.to_numeric(df['account_Charges_Total'], errors='coerce')
print("Number of values converted to NaN in 'account_Charges_Total': ", df['account_Charges_Total'].isnull().sum())
```

```python
df['Contas_Diarias'] = (df['account_Charges_Monthly'] / 30).round(2)
df.head()
```

#ðŸ“Š Carga e anÃ¡lise

```python
print(f"MÃ©dia de churns: {df['Churn'].mean() * 100:.2f}%")
print(f"MÃ©dia de gastos p/ pessoa no mes: {df['account_Charges_Monthly'].mean():.2f}")
print(f"Mediana de gastos p/ pessoa no mes: {df['account_Charges_Monthly'].median():.2f}")
print(f"Desvio PadrÃ£o de gastos p/ pessoa no mes: {df['account_Charges_Monthly'].std():.2f}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 5))
plt.pie(df['Churn'].value_counts(), labels=['NÂº de clientes que saÃ­ram', 'NÂº de clientes que permaneceram'], autopct='%1.1f%%', colors=['#ff4a1c', '#b2ffa9'])
plt.title('DistribuiÃ§Ã£o de Churn')
plt.show()
```

```python
categorical_cols = ['customer_gender', 'account_Contract', 'account_PaymentMethod', 'customer_Partner', 'customer_Dependents']

for col in categorical_cols:
    plt.figure(figsize=(7, 4))
    sns.countplot(data=df, x=col, hue='Churn', palette=['#ff4a1c', '#b2ffa9'])
    plt.ylabel('Quantidade')
    plt.title(f'DistribuiÃ§Ã£o de saÃ­das por {col}')
    plt.legend(title='Churn', loc='upper right', labels=['Saiu', 'NÃ£o Saiu'])
    if col == 'customer_Partner':
        plt.xticks(ticks=[0, 1], labels=['Tem Esposa', 'NÃ£o tem esposa'])
    elif col == 'customer_Dependents':
        plt.xticks(ticks=[0, 1], labels=['Tem dependentes', 'NÃ£o tem dependentes'])
    elif col == 'account_PaymentMethod':
        plt.xticks(rotation=15)
    plt.show()
    print()
```
